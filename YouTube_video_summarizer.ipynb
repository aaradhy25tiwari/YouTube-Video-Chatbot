```python
from youtube_transcript_api import YouTubeTranscriptApi
import urllib.parse as urlparse
```

```python
def get_video_id(url):
    """Extracts the video ID from various YouTube URL formats."""
    parsed_url = urlparse.urlparse(url)
    if parsed_url.hostname == 'youtu.be':
        return parsed_url.path[1:]
    if parsed_url.hostname in ('www.youtube.com', 'youtube.com'):
        if parsed_url.path == '/watch':
            p = urlparse.parse_qs(parsed_url.query)
            return p.get('v', [None])[0]
    return url # Fallback if it's already just the ID
```

```python
url = input("Enter the YouTube video URL: ")
video_id = get_video_id(url)
```

```python
print(video_id)
```

```python
# Create the instance
api = YouTubeTranscriptApi()

# Fetch specifically using the 'hi' (Hindi) language code
# The library takes a list, so we use ['hi']
transcript = api.fetch(video_id, languages=['en', 'hi'])

print(transcript)
```

```python
transcript_text = " ".join([line.text for line in transcript])

print(transcript_text)
```

```python
from deepmultilingualpunctuation import PunctuationModel

# This model supports English, German, French, Spanish, Italian and others.
# It handles multilingual context much better than rpunct.
model = PunctuationModel()
clean_text = model.restore_punctuation(transcript_text)

print(clean_text)
```

```python
from google import genai
from google.colab import userdata
```

```python
try:
    api_key = userdata.get('GEMINI_API_KEY')
    client = genai.Client(api_key=api_key)
    print("✅ Gemini Client initialized successfully!")
except Exception as e:
    print("❌ Secret 'GEMINI_API_KEY' not found in Colab Secrets.")
```

```python

prompt = f"Summarize this text: \ntext = {transcript_text}"
```

```python
# Use the stable production model from your list
response = client.models.generate_content(
    model="gemini-2.5-flash-lite",
    contents=prompt,
    config={
        "temperature": 1.0,
        "max_output_tokens": 1024, # Increased for your long Gaurav Tiwari story
        "top_p": 0.95,
    }
)

print(response.text)
```

```python
# --- Start of Q&A Section ---

print("\n--- 🤖 Video Q&A Mode ---")
print("Ask anything about the video (type 'exit' to stop):")

# We initialize a chat session and provide the transcript as context in the history
chat = client.chats.create(
    model="gemini-2.5-flash-lite",
    config={
        "system_instruction": f"You are a helpful assistant. Use the following video transcript to answer questions accurately: {clean_text}"
    }
)

while True:
    user_query = input("\nUser: ")

    if user_query.lower() in ['exit', 'quit', 'bye']:
        print("Exiting Q&A mode. Goodbye!")
        break

    try:
        # Generate the response based on the transcript context
        qa_response = chat.send_message(user_query)
        print(f"\nGemini: {qa_response.text}")
    except Exception as e:
        print(f"An error occurred: {e}")

# --- End of Q&A Section ---
```
